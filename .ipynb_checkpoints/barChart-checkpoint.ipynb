```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
company = ['Google', 'Amazon', 'Microsoft', 'Meta']
revenue = [105, 135, 126, 85]
profit = [40, 12, 34, 15]
bar_width = 0.34

x = range(len(company))  # Generate x-axis positions for the bars

plt.xlabel('Companies')
plt.ylabel('Revenue and Profit (in Bn)')  # Updated the ylabel to reflect both revenue and profit
plt.title('US Tech Stocks')
plt.bar([i - bar_width/2 for i in x], revenue, width=bar_width, label='Revenue')  # Adjusted the x position for revenue bars
plt.bar([i + bar_width/2 for i in x], profit, width=bar_width, label='Profit')  # Adjusted the x position for profit bars
plt.xticks(x, company)  # Set x-axis labels to company names
plt.legend()
plt.grid()
plt.show()
```
