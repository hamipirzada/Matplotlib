```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men = [113, 85, 90, 150, 145, 88, 93, 115, 135, 80, 77, 82, 136]
blood_sugar_women = [110, 80, 95, 130, 125, 95, 110, 90, 140, 100, 97, 98, 86]
plt.xlabel('Sugar Range')
plt.ylabel('Total no. of Patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar,blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['g','y'], label=['Men', 'Women'])
plt.legend()
```
