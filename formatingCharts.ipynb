```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [25,41,32,40,45,36,32]
```

```python
# g+ means green color with + sign instead of line
plt.plot(x,y, 'g+')
```

```python
plt.plot(x,y, '--')
```

```python
plt.plot(x,y, '.')
```

```python
plt.plot(x,y, 'D-')
```

```python
plt.plot(x,y, 'rD')
```

```python
# Using alpha property to control the transparency of the line
# Range of alpha is from 0-1 (0 means transparent and 1 means not transparent(opaque))
plt.plot(x, y, color='green', alpha=0.5, animated=True)
```
